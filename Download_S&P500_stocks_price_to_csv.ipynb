```python
import bs4 as bs
import pickle
import requests
import datetime as dt
import os
import pandas as pd
import pandas_datareader.data as web

# Pretend use browser
headers = {'User-Agent': 'Mozilla/75.0 Chrome/80.0.3987.163 '}
```

```python
#  stock symbol
def save_sp500_symbols():
    
    # crawler websites
    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies',headers = headers)

    # source is text, parser is lxml
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    
    # table
    table = soup.find_all("table",{"id" : "constituents"})[0]
    
    symbols = []
    
    #tr is row of table
    for row in table.findAll('tr')[1:]:
        
        # soup object get text)
        symbol = row.findAll('td')[0].get_text(strip=True)

        symbols.append(symbol)
        
    #  save result to pickle
    with open("SP500.pickle","wb") as file:
        pickle.dump(symbols, file)
    
    return symbols
```

```python
save_sp500_symbols()
```

```python
# save price from spider
def get_data_from_datareader(beg_y, beg_m, beg_d, end_y, end_m, end_d, reload_symbols=False):

    # if reload, execute save_sp500_symbols()
    if reload_symbols:
        symbols = save_sp500_symbols()
    # if no reload, read SP500.pickle     
    else:
        with open("SP500.pickle","rb") as file:
            symbols = pickle.load(file)
    
    # if no folder exist, mkdir
    if not os.path.exists('SP500'):
        os.makedirs('SP500')
        
    beg = dt.datetime(beg_y, beg_m, beg_d)
    end = dt.datetime(end_y, end_m, end_d)
#     print(beg,end)

    #for loop price from 'stooq'
    for symbol in symbols:
        #try: check symbol is downloaed
        try:
            if not os.path.exists('SP500/{}.csv'.format(symbol)):
                df = web.DataReader(symbol, 'stooq', beg, end)
                df.to_csv('SP500/{}.csv'.format(symbol))
            else:
                print('already have {}'.format(symbol))
        # except
        except:
            print(symbol + "can not download")
    print("crawler done")
```

```python
get_data_from_datareader(2020,1,1,2020,4,17)
```
