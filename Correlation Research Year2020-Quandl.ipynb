```python
import quandl
import seaborn as sns
import pandas as pd
```

```python
begdate="2020-01-01"
enddate="2020-03-07"
```

```python
#USD https://www.quandl.com/data/CHRIS/ICE_DX1-US-Dollar-Index-Futures-Continuous-Contract
USD = quandl.get("CHRIS/ICE_DX1", start_date=begdate, end_date=enddate)
USD = USD.rename(columns={"Settle": "USD",})
print(USD.head(2))
print(USD.tail(1))
```

```python
#Crude Oil Futures, Continuous Contract #1 (CL1) (Front Month) 
#https://www.quandl.com/data/CHRIS/CME_CL1-Crude-Oil-Futures-Continuous-Contract-1-CL1-Front-Month
Oil = quandl.get("CHRIS/CME_CL1", start_date=begdate, end_date=enddate)
Oil = Oil.rename(columns={"Settle": "Oil",})
print(Oil.head(2))
print(Oil.tail(1))
```

```python
#Gold Prices (Daily) - Currency USD 
#https://www.quandl.com/data/WGC/GOLD_DAILY_USD-Gold-Prices-Daily-Currency-USD
GOLD = quandl.get("WGC/GOLD_DAILY_USD", start_date=begdate, end_date=enddate)
GOLD = GOLD.rename(columns={"Value": "GOLD",})
print(GOLD.head(2))
print(GOLD.tail(1))
```

```python
#2007!!!
#E-mini Dow ($5) Futures, Continuous Contract #1 (YM1) (Front Month)
#https://www.quandl.com/data/CHRIS/CME_YM1-E-mini-Dow-5-Futures-Continuous-Contract-1-YM1-Front-Month
DJF = quandl.get("CHRIS/CME_YM1", start_date=begdate, end_date=enddate)
DJF = DJF.rename(columns={"Settle": "DJF",})
print(DJF.head(2))
print(DJF.tail(1))
```

```python
#S&P 500 Futures, Continuous Contract #1 (SP1) (Front Month)
#https://www.quandl.com/data/CHRIS/CME_SP1-S-P-500-Futures-Continuous-Contract-1-SP1-Front-Month#https://www.quandl.com/data/CHRIS/CME_YM1-E-mini-Dow-5-Futures-Continuous-Contract-1-YM1-Front-Month
SPF = quandl.get("CHRIS/CME_SP1", start_date=begdate, end_date=enddate)
SPF = SPF.rename(columns={"Settle": "SPF",})
print(SPF.head(2))
print(SPF.tail(1))
```

```python
#U.S. Treasury Bond Futures, Continuous Contract #1 (US1) (Front Month)
#https://www.quandl.com/data/CHRIS/CME_US1-U-S-Treasury-Bond-Futures-Continuous-Contract-1-US1-Front-Month
TYF = quandl.get("CHRIS/CME_US1", start_date=begdate, end_date=enddate)
TYF = TYF.rename(columns={"Settle": "TYF",})
print(TYF.head(2))
print(TYF.tail(1))
```

```python
# data = SPY.Close WIP.Close USO.close
data =  pd.concat([SPF.SPF, Oil.Oil,GOLD.GOLD,USD.USD,TYF.TYF], axis=1)
# data =  data.rename(columns={0: "SPY", 1: "WIP"})
print(data.head(5))
```

```python
daily_returns = data.pct_change()
# daily_returns = data
correlations = daily_returns.corr()
correlations
```

```python
# begdate="2020-01-01" enddate="2020-03-07"
# Quandll data
# SPF:S&P500 continue future,Oil: Crude Oil,USD:USD Dollar Index;TYF:10 year treasury bond continue future
# Nega tive Correlation : (SPF,GOLD),(SPF,TYF),(Oil,TYF),(Gold,USD),(USD,TYF)
```

```python
sns.heatmap(correlations, annot = True)
```
